## Imports
Import numpy, tensorflow and keras

```
import keras
#import tensorflow as tf
import numpy as np
#from tensorflow import keras
from keras.callbacks import TensorBoard
```

## Define and Compile neural network
1 layer and 1 neuron per layer, input just 1 value

```
model = keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
```

We need to define a loss function and a optiomizer.

We use 'MEAN SQUARED ERROR' for the loss and 'STOCHASTIC GRADIENT DESCENT' for the optimizer

```
model.compile(optimizer='sgd', loss='mean_squared_error')
tensor_board = TensorBoard(log_dir='/home/wayne/prog/Projects/DL-Study/logs1/Linear')
```

## Data
Use numpy to provide lots of array type data structures.

```
xs = np.array([-1.0,  0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=float)
```

## Training the Neural Network
Use **model.fit** call to learn the relationship between **xs** and **ys**.
Training for 500 epochs

```
model.fit(xs, ys, epochs=500, verbose=1, callbacks=[tensor_board])
```

```
print(model.predict([10.0]))
```
